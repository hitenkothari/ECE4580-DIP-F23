```python
from google.colab import drive
from skimage import io,data,restoration
from skimage import color
import numpy as np
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

PART I: Weiner Filter

```python
image = io.imread("/content/drive/MyDrive/Images/smeared.png") #reading the grayscaled image
kernel_size=72 #value selected after multiple iterations
balance = 5000 #value selected after multiple iterations
psf = np.ones((1, kernel_size)) / 1*kernel_size #since the degradation is only in horizontal direction, 1d matrix required
restored_image = restoration.wiener(image, psf, balance) #applying weiner restoration

#plots
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Smeared Image')
plt.axis('off')
plt.subplot(2, 1, 2)
plt.imshow(restored_image, cmap='gray')
plt.title('Image Restored using Weiner Filter')
plt.axis('off')
plt.show()
```

PART II: Hue Variation

```python
image = io.imread("/content/drive/MyDrive/Images/bananaplant.png") #loading image

hsv_image= color.rgb2hsv(image) #rgb to hsv

br1,br2,bc1,bc2 = 120,950,140,800 #row and column limits of banana

condition = (hsv_image[br1:br2,bc1:bc2,0] >= 20/360) & (hsv_image[br1:br2,bc1:bc2,0] <= 60/360) #values arrived after trial and error to minimize background change
hsv_image[br1:br2,bc1:bc2,0][condition] +=180/360 #converting to blue

image3 = color.hsv2rgb(hsv_image) #converting hsv to rgb

#plots
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.subplot(2, 1, 2)
plt.imshow(image3)
plt.title('Output Image')
plt.axis('off')
plt.show()
```
